```python
import yfinance as yf
import pandas as pd

# 1. Create a Ticker object for Tesla (TSLA)
tesla = yf.Ticker("TSLA")

# 2. Download historical stock data for the maximum period
tesla_data = tesla.history(period="max")

# 3. Reset the index to make 'Date' a regular column
tesla_data.reset_index(inplace=True)

# 4. Display the first five rows to verify the extraction
print("First 5 rows of Tesla Stock Data:")
print(tesla_data.head())
```

```python
import pandas as pd

# The URL provided for the Tesla financial data in the course lab
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
# 1. Use pandas to read all HTML tables from the URL
tables = pd.read_html(url)
# 2. Assume the quarterly revenue table is the second one (index 1) on the page
tesla_revenue = tables[1]
# 3. Rename columns for clarity and consistency
tesla_revenue.columns = ['Date', 'Revenue']
# 4. Clean the 'Revenue' column by removing '$' and ',' characters
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '')
# 5. Drop any rows where 'Revenue' or 'Date' is a missing value (NaN)
tesla_revenue.dropna(inplace=True)
# 6. Drop any rows where the 'Revenue' column is an empty string
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
# 7. Display the last five rows of the cleaned revenue data
print("Last 5 rows of Tesla Revenue Data:")
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import pandas as pd

# 1. Create a Ticker object for GameStop (GME)
gamestop = yf.Ticker("GME")

# 2. Download historical stock data for the maximum period
gme_data = gamestop.history(period="max")

# 3. Reset the index to make 'Date' a regular column
gme_data.reset_index(inplace=True)

# 4. Display the first five rows to verify the extraction
print("First 5 rows of GameStop Stock Data:")
print(gme_data.head())
```

```python
import pandas as pd
# The URL provided for the GameStop financial data in the course lab
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# 1. Use pandas to read all HTML tables from the URL
tables_gme = pd.read_html(url_gme)

# 2. Assume the quarterly revenue table is the second one (index 1) on the page
gme_revenue = tables_gme[1]

# 3. Rename columns for clarity and consistency
gme_revenue.columns = ['Date', 'Revenue']

# 4. Clean the 'Revenue' column by removing '$' and ',' characters
gme_revenue["Revenue"] = gme_revenue['Revenue'].str.replace('$', '').str.replace(',', '')

# 5. Drop any rows where 'Revenue' or 'Date' is a missing value (NaN)
gme_revenue.dropna(inplace=True)

# 6. Drop any rows where the 'Revenue' column is an empty string
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]

# 7. Display the last five rows of the cleaned revenue data
print("Last 5 rows of GameStop Revenue Data:")
print(gme_revenue.tail())
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd # Import just in case it wasn't done in the same cell as the function

# Define the function to create the dashboard
def make_graph(stock_data, revenue_data, stock):
    # Ensure Date columns are datetime objects
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    revenue_data['Date'] = pd.to_datetime(revenue_data['Date'])
    
    # Create subplots: 2 rows, 1 column, with shared x-axis
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, 
                        subplot_titles=(f"{stock} Stock Price History", f"{stock} Quarterly Revenue"), 
                        vertical_spacing=0.1)

    # Plot Stock Price (Close) on the top subplot
    fig.add_trace(go.Scatter(x=stock_data.Date, y=stock_data.Close, name="Stock Price"), row=1, col=1)
    
    # Plot Revenue on the bottom subplot
    # Note: Revenue data needs to be converted to numeric for plotting
    revenue_data['Revenue'] = pd.to_numeric(revenue_data['Revenue'])
    fig.add_trace(go.Scatter(x=revenue_data.Date, y=revenue_data.Revenue, name="Revenue"), row=2, col=1)

    # Update axis labels and layout
    fig.update_xaxes(title_text="Date")
    fig.update_yaxes(title_text="Stock Price (USD)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue (Millions USD)", row=2, col=1)
    fig.update_layout(title_text=f"Financial Dashboard for {stock}", 
                      showlegend=True, 
                      height=900)
    
    fig.show()

# Call the function for Tesla
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
# Call the make_graph function for GameStop (GME)
make_graph(gme_data, gme_revenue, 'GameStop')
```
